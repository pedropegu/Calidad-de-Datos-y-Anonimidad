## Limpieza de los datos

Eliminamos columnas innecesarias como el √≠ndice y valores duplicados.

```python
df = df.drop(['index'], axis=1)
df.drop_duplicates(inplace=True)
df
```

Analizamos la columna "Company Name" donde encontramos saltos de l√≠nea junto a ciertos n√∫meros.

```python
df['Company Name']
```

Seleccionamos la parte que nos interesa que se trata del nombre.

```python
df['Company Name'] = df['Company Name'].map(lambda x: x.split('\n')[0])
df.head()
```

Remplazamos valores como -1 o Unknown a NA el cual ser√° el valor que reprentar√° a los valores nulos.

```python
categorical_cols = ['Job Title', 'Salary Estimate', 'Job Description', 
                    'Company Name', 'Location', 'Headquarters', 'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'Competitors']

df[categorical_cols] = df[categorical_cols].replace('-1', 'NA')
df[categorical_cols] = df[categorical_cols].replace('Unknown', 'NA')
df['Rating'] = df['Rating'].replace(-1., 0)

df.head()
```

Verificamos el estado de la columna Salary Estimate donde se observa un rango de valores, para limpiar este campo separaremos el rango y crearemos columnas para el valor de rango m√≠nimo, m√°ximo y para la media.

```python
df['Salary Estimate'] = df['Salary Estimate'].map(lambda x: x.split('(')[0])

# ELIMINAR LOS RANGOS Y CREAR COLUMNAS CORRESPONDIENTES PARA EL MIN Y EL MAX

df['Min_Salary'] = df['Salary Estimate'].map(lambda x: int(x.split('-')[0].strip('$').replace('K',''))*1000)
df['Max_Salary'] = df['Salary Estimate'].map(lambda x: int(x.split('-')[1].strip('$').replace('K',''))*1000)

df['Avg_Salary'] = (df['Min_Salary'] + df['Max_Salary']) / 2
```

```python
df.head()
```

```python
df = df.drop(['Salary Estimate'], axis=1)
df.head()
```

Atendiendo a la fecha de fundaci√≥n de la empresa, este campo simplemente recibir√° una limpieza de valores nulos, adem√°s a√±adiremos una columna extra denominada "Company_Age" en la cual tendremos la antig√ºedad de la empresa desde el a√±o 2024.

```python
conteo_valores = df['Founded'].unique()
print(conteo_valores)


df['Founded'] = df['Founded'].map(lambda x: x if x > 0 else 0)
df['company_age'] = df['Founded'].map(lambda x: x if x<1 else 2024-x)
conteo_valores = df['Founded'].unique()
print(conteo_valores)
```

```python
df[["Founded","company_age"]].head()
```

Limpiaremos los campos de localizaci√≥n y sede central, crearemos un campo same_state que se obtendr√° de la siguiente verificaci√≥n: sede central = localizaci√≥n. Si esta se cumple ser√° 1. Tamb√≠en renombramos el campo de localizaci√≥n y lo dejaremos en Job_State.

```python
df['Location'] = df['Location'].map(lambda x: x.split(',')[0] if ',' in x else 'NA')
df['Headquarters'] = df['Headquarters'].map(lambda x: x.split(',')[0] if ',' in x else 'NA')

df['job_state'] = df['Location']
df['same_state'] = df.apply(lambda x: 1 if x.Location==x.Headquarters else 0,axis=1)

df.drop(['Location'], axis=1, inplace=True)
df
```

Respecto al t√≠tulo del trabajo, este sufrir√° una peque√±a limpieza donde nos quedaremos con lo m√°s importante del contenido obviando aquello que es innecesario.

```python
df['Job Title'].value_counts()
```

```python
df['Job Title'].unique()
```

```python
def job_title_classifier(title):

    if 'data scientist' in title.lower():
        return 'Data scientist'
    elif 'data engineer' in title.lower():
        return 'Data engineer'
    elif 'analyst' in title.lower():
        return 'Analyst'
    elif 'machine learning' in title.lower():
        return 'MLE'
    elif 'manager' in title.lower():
        return 'Manager'
    elif 'director' in title.lower():
        return 'Director'
    else:
        return 'NA'
    
df['Job_Title_cleaned'] = df['Job Title'].apply(job_title_classifier)
df['Job_Title_cleaned'].value_counts()
```

Adem√°s, crearemos un campo llamado job_level que corresponder√° a la experiencia requerida (Este campo lo obtendremos de la columna Job Title).

```python
def Level_classifier(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr.' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'Senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'Junior'
    else:
        return 'NA'
    
df['Job_Level'] = df['Job Title'].apply(Level_classifier)
df['Job_Level'].value_counts()
```

```python
df.drop(['Job Title'], axis=1, inplace=True)
df["Job_Title"] = df["Job_Title_cleaned"]
df.drop(["Job_Title_cleaned"], axis=1, inplace=True)
df.head()
```

Comprobamos los valores existentes en el rating y sustituimos valores como los negativo ya que no encajan en este campo. 

```python
df['Rating'].unique()
```

```python
df['Rating'] = np.where(df.Rating < 0, 0, df.Rating)
df['Rating'].unique()
```

El campo de descripci√≥n desaparecer√° por completo, sin embargo crearemos campos por cada aptitud que se deba necesitar en este. Las aptitudes las podemos establecer echando un vistazo sobre las descripciones, tambi√©n sabiendo el contexto del dataset puede resultar sencillo obtenerlas.

```python
df['Job Description'].unique()
```

```python
def check_skill(skill, description):
    return 1 if skill.lower() in description.lower() else 0

skills_to_check = ['python','excel','sql','tableau','aws','spark','hadoop','azure']

for skill in skills_to_check:
    df[skill.upper()] = df['Job Description'].apply(lambda x: check_skill(skill, x))
```

```python
df.drop(['Job Description'], axis=1, inplace=True)
```

Para finalizar, hemos decidido convertir el nombre de las columnas en min√∫sculas para que todos los nombres est√©n formateados. Adem√°s ordeno las columnas para que este se pueda observar mejor.

```python
def columnas_en_mayusculas(df):
    df.columns = map(str.lower, df.columns)
    return df

df = columnas_en_mayusculas(df)
orden_columnas = ['rating', 'company name','headquarters', 'founded', 'company_age',
                'type of ownership', 'industry', 'sector', 'revenue', 'competitors',
                'min_salary', 'max_salary', 'avg_salary', 'job_state', 'same_state',
                'job_title', 'job_level', 'python', 'excel', 'sql', 'tableau',
                'aws', 'spark', 'hadoop', 'azure', 'segment']

df = df[orden_columnas]
df.to_csv('../Data/Cleaned_Data.csv', index=False)
```
